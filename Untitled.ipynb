```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import os
import csv
from tensorflow.python.framework import ops
ops.reset_default_graph()
```

```python
# same params and variables initialization as log reg.
W = tf.Variable(tf.random_normal(shape=[7,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))
# former inference is now used for combining inputs
```

<div style="direction:rtl">
  :        .         .

</div>
\begin{equation}
\sum_{n=1}^{7}A_n.x_n +b
\end{equation}


```python
def combine_inputs(X):
    return tf.matmul(X, W) + b
# new inferred value is the sigmoid applied to the former
```

<div style="direction:rtl">
       sigmoid    .  
</div>

```python
def inference(X):
    return tf.sigmoid(combine_inputs(X))
```


<div style="direction:rtl">
    
### Loss Function
</div>
<div style="direction:rtl">
              

</div>



```python
def loss(X, Y):
    return tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=tf.squeeze(combine_inputs(X)), labels=Y))
```

```python
def inputs():
    birth_weight_file = 'birth_weight.csv'
    birth_data = []
    with open(birth_weight_file, newline='') as csvfile:
        csv_reader = csv.reader(csvfile)
        birth_header = next(csv_reader)
        for row in csv_reader:
            birth_data.append(row)
    birth_data = [[float(x) for x in row] for row in birth_data]
        # Pull out target variable
    y_vals = np.array([x[0] for x in birth_data])
        # Pull out predictor variables (not id, not target, and not birthweight)
    x_vals = np.array([x[1:8] for x in birth_data])
    seed = 99
    np.random.seed(seed)
    tf.set_random_seed(seed)
    # Split data into train/test = 80%/20%
    train_indices = np.random.choice(len(x_vals), round(len(x_vals)*0.8), replace=False)
    test_indices = np.array(list(set(range(len(x_vals))) - set(train_indices)))
    x_vals_train = x_vals[train_indices]
    x_vals_test = x_vals[test_indices]
    y_vals_train = y_vals[train_indices]
    y_vals_test = y_vals[test_indices]
    return tf.to_float(x_vals_train),tf.to_float(y_vals_train),tf.to_float(x_vals_test),tf.to_float(y_vals_test)
```

<div style="direction:rtl">
           
</div>

`0.01`

```python
def train(total_loss):
    learning_rate = 0.01
    return tf.train.GradientDescentOptimizer(learning_rate).minimize(total_loss)
```

<div style="direction:rtl">
             .                        Accuracy   
</div>


```python
def evaluate(sess, X, Y):
    prediction = tf.round(inference(X))
    predictions_correct = tf.cast(tf.equal(prediction, Y), tf.float32)
    accuracy = tf.reduce_mean(predictions_correct)
    return sess.run(accuracy)
```

<div style="direction:rtl">
                              
</div>


```python
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    x_vals_train ,  y_vals_train,x_vals_test,y_vals_test = inputs()
    total_loss = loss(x_vals_train, y_vals_train)
    train_op = train(total_loss)
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)
    # actual training loop
    training_steps = 1000
    for step in range(training_steps):
        sess.run([train_op])
        if step % 100 == 0:
            print ("\nloss: ", sess.run([total_loss]))
            print("\nTriain ACC:",evaluate(sess, x_vals_train ,  y_vals_train))
            print("\nTest ACC:",evaluate(sess, x_vals_test,y_vals_test))
            print("-"*20)
            
            
    coord.request_stop()
    coord.join(threads)

```
