# طبقه بند دودویی Logistic Regression:

در بسیاری از موارد مقدار متغییر هدف که قرار است پیش بینی شود یک مقدار گسسته است در حالی که در رگرسیون هدف پیش بینی یک مقدار در یک بازه پیوسته بود در این مساله پیش بینی براساس دسته های محدودی صورت می گیرد که به این مسال اصطلاحا مسال طبقه بندی گفته می شود. اگر متغییر هدف تنها دارای دو دسته 0 یا منفی و 1 یا مثبت باشد آن را طبقه بند دودویی یا باینری می گویند. در این بخش هدف ما این است که با استفاده از خطی که رگرسیون خطی بر روی داده ها یاد می گیرد طبقه بندی دودویی را انجام دهیم. رابطه رگرسیون خطی به صورت زیر است.

$$ y = Ax + b$$

هدف ما  استفاده از این خط است که  در آن وسط خط یعنی 0.5 را حد آستانه در نظر می گیریم.

</div>

<div style="direction:rtl">
هدف ما  استفاده از این خط است وسط خط یعنی 0.5 را حد آستانه در نظر می گیریم. این بدین معنی است که هرچیزی بالای حد آستانه باشدکلاس 1 است و درغیر این صورت کلاس آن 0 است. شکل زیر نمونه ای از حد آستانه را نشان می دهد.
</div>




<div style="direction:rtl">
بسته به پراکندگی داده ها موقعیت حد استانه ممکن است تغییر کند ولی در همه حال مقدار حد استانه را برابر 0.5 در نظر می گیرند.
</div>



<div style="direction:rtl">

 مشکل اساسی که در رگرسیون برای این کار وجود دارد این است که محدوده  در آن کنترل نمی شود یعنی مقدار متغییر هدف می تواند به صورت زیر تغییر کند.
 
$y>1$

$y<0$

که این مقدار برای پیش بینی داده نیاز است که محدود شود برای این منظور از تابع Sigmoid استفاده می کنیم. در این صورت تابع رگرسیون به صورت زیر تغییر می کند.

$ \textbf{y} = sigmoid(\textbf{A}\times \textbf{x} + \textbf{b})$
</div>

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import os
import csv
from tensorflow.python.framework import ops
ops.reset_default_graph()
```

<div style="direction:rtl">
در ابتدا دو متغییر برای مدل رگرسیون تعریف می کنیم. این دو متغییر قرار است طی فرایند یادگیری تغییر کنند و مقدار بهینه را پیدا کنند 
</div>


```python
# same params and variables initialization as log reg.
W = tf.Variable(tf.random_normal(shape=[7,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))
# former inference is now used for combining inputs
```

<div style="direction:rtl">
تعریف تابع رگرسیون: برای محاسبه رگرسیون تابعی را تعریف می کنیم. این تابع قرار است مقدار زیر را حساب کند.

</div>
\begin{equation}
\sum_{n=1}^{7}A_n.x_n +b
\end{equation}


```python
def combine_inputs(X):
    return tf.matmul(X, W) + b
# new inferred value is the sigmoid applied to the former
```

<div style="direction:rtl">
بر روی نتیجه خروجی رگرسیون تابع  sigmoid را اعمال می کنیم.  
</div>

```python
def inference(X):
    return tf.sigmoid(combine_inputs(X))
```


<div style="direction:rtl">
    
### Loss Function
</div>
<div style="direction:rtl">
برای تابع هزینه نیز می توان از هر یک از توابع بررسی شده استفاده کرد

</div>



```python
def loss(X, Y):
    return tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=tf.squeeze(combine_inputs(X)), labels=Y))
```

```python
import pandas as pd
data=pd.read_csv('birth_weight.csv')
data.head()
```

<div style="direction:rtl">
داده هارا فراخوانی می کنیم و 80 درصد از آن ها را به عنوان داده ی آموزشی و 20 درصد را داده ی تست در نظر می گیرم.
</div>

```python
def inputs():
    birth_weight_file = 'birth_weight.csv'
    birth_data = []
    with open(birth_weight_file, newline='') as csvfile:
        csv_reader = csv.reader(csvfile)
        birth_header = next(csv_reader)
        for row in csv_reader:
            birth_data.append(row)
    birth_data = [[float(x) for x in row] for row in birth_data]
        # Pull out target variable
    y_vals = np.array([x[0] for x in birth_data])
        # Pull out predictor variables (not id, not target, and not birthweight)
    x_vals = np.array([x[1:8] for x in birth_data])
    seed = 99
    np.random.seed(seed)
    tf.set_random_seed(seed)
    # Split data into train/test = 80%/20%
    train_indices = np.random.choice(len(x_vals), round(len(x_vals)*0.8), replace=False)
    test_indices = np.array(list(set(range(len(x_vals))) - set(train_indices)))
    x_vals_train = x_vals[train_indices]
    x_vals_test = x_vals[test_indices]
    y_vals_train = y_vals[train_indices]
    y_vals_test = y_vals[test_indices]
    return tf.to_float(x_vals_train),tf.to_float(y_vals_train),tf.to_float(x_vals_test),tf.to_float(y_vals_test)
```

<div style="direction:rtl">
مقدار نرخ یادگیری را در تابع بهینه ساز  تنظیم می کنیم
</div>

`0.01`

```python
def train(total_loss):
    learning_rate = 0.01
    return tf.train.GradientDescentOptimizer(learning_rate).minimize(total_loss)
```

<div style="direction:rtl">
پس از این که مدل را تعریف کردیم باید بتوانیم آن را ارزیابی کنیم. برای ارزیابی آن داده های تست و داده های آموزش را به پارامترهای یادگیری می فرستیم و نتیجه ی خروجی را براساس معیار Accuracy بیان می کنیم
</div>


```python
def evaluate(sess, X, Y):
    prediction = tf.round(inference(X))
    predictions_correct = tf.cast(tf.equal(prediction, Y), tf.float32)
    accuracy = tf.reduce_mean(predictions_correct)
    return sess.run(accuracy)
```

<div style="direction:rtl">
در این بخش نیز می توان داده ها را هم بصورت دسته ای و هم به صورت کلی به مدل داده که ما از حالت غیر دسته ای استفاده میکنیم 
</div>


```python
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    x_vals_train ,  y_vals_train,x_vals_test,y_vals_test = inputs()
    total_loss = loss(x_vals_train, y_vals_train)
    train_op = train(total_loss)
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)
    # actual training loop
    training_steps = 1000
    for step in range(training_steps):
        sess.run([train_op])
        if step % 100 == 0:
            print ("\nloss: ", sess.run([total_loss]))
            print("\nTriain ACC:",evaluate(sess, x_vals_train ,  y_vals_train))
            print("\nTest ACC:",evaluate(sess, x_vals_test,y_vals_test))
            print("-"*20)
            
            
    coord.request_stop()
    coord.join(threads)

```
